```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books = pd.read_csv('/books.csv')
books
```

```python
books.columns
```

```python
books = books[['id', 'book_id','isbn',
       'isbn13', 'authors','title','average_rating', 'ratings_count','image_url',
       'small_image_url']]
```

```python
book_tags = pd.read_csv('/book_tags.csv')
book_tags
```

```python
book_tags = book_tags.rename(columns={'goodreads_book_id':'book_id'})
```

```python
tags = pd.read_csv('/tags.csv')
tags
```

```python
tagged_books = pd.merge(book_tags, tags, on='tag_id', how='inner')
tagged_books
```

```python
new_tagged_books = tagged_books.groupby('book_id')['tag_name'].agg(lambda x: list(x)).reset_index()
new_tagged_books
```

```python
books_with_tags = pd.merge(books,new_tagged_books,on='book_id',how='inner')
books_with_tags
```

```python
books_with_tags['tags'] = books_with_tags.apply(lambda row: [row['authors']] + row['tag_name'], axis=1)
books_with_tags
```

```python
books_with_tags = books_with_tags.drop(columns=['tag_name'],axis=1)
```

```python
books_with_tags['tags'] = books_with_tags['tags'].apply(
    lambda tag_list: [tag.lower().replace(' ', '').replace('-', '').replace('.','').replace(',',' ') for tag in tag_list]
)
books_with_tags
```

```python
books_with_tags['tags'] = books_with_tags['tags'].apply(lambda x: " ".join(x))
books_with_tags
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = vectorizer.fit_transform(books_with_tags['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(vectors)
similarity_score
```

```python
list(enumerate(similarity_score))
```

```python
def rec(book_name):
  index = books[books['title'] == book_name].index[0]
  similar_books = similarity_score[index]
  similar_books = sorted(list(enumerate(similar_books)), key=lambda x: x[1], reverse=True)
  recs = []
  for book in similar_books:
    recs.append(book[0])
  recommendations = books_with_tags.loc[recs]
  return recommendations[1:11]
```

```python
rec('The Mysterious Island (Extraordinary Voyages, #12)')
```
